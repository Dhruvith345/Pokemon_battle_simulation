```python
def moderator_calculation(pok1_type1, pok1_type2, pok2_type1, pok2_type2):
  moderator = 0
  for i in [pok1_type1, pok1_type2]:
    if i == 'Null':
      break
    multiplier = 1
    for j in [pok2_type1, pok2_type2]:
      if j == 'Null':
        break
      if pokemon_moderator[i][j] == 0:
        break
      multiplier *= pokemon_moderator[i][j]
    if moderator < multiplier:
      moderator = multiplier
  return moderator
```

```python
def battle_simulation(attack, defense, extra_attack, modulator):
    return ((10 * (attack / defense)) + extra_attack*3) * modulator
```

```python

def battle_stats_outcome_extractor(pokemon1, pokemon2):
    pokemon1_data = pokemon_dataframe[pokemon_dataframe['Name'] == pokemon1].iloc[0]
    pokemon2_data = pokemon_dataframe[pokemon_dataframe['Name'] == pokemon2].iloc[0]
    pokemon1_type1, pokemon1_type2 = pokemon1_data['Type 1'], pokemon1_data['Type 2']
    pokemon2_type1, pokemon2_type2 = pokemon2_data['Type 1'], pokemon2_data['Type 2']
    attack1, attack2 = pokemon1_data['Attack'], pokemon2_data['Attack']
    hp1, hp2 = pokemon1_data['HP'], pokemon2_data['HP']
    defense1, defense2 = pokemon1_data['Defense'], pokemon2_data['Defense']
    speed1, speed2 = pokemon1_data['Speed'], pokemon2_data['Speed']
    extra_attack1 = pokemon2_data['Attack'] - pokemon1_data['Defense']
    extra_attack2 = pokemon1_data['Attack'] - pokemon2_data['Defense']

    if extra_attack1 <0:
        extra_attack1 = 0
    if extra_attack2 <0:
        extra_attack2 = 0

    stats1 = attack1 + defense1 + speed1 + hp1 + extra_attack1
    stats2 = attack2 + defense2 + speed2 + hp2 + extra_attack2
    if(speed1!=speed2):
      turn = 1 if speed1 > speed2 else 0
      while hp1 > 0 and hp2 > 0:
        if turn != 0:
            modulator= moderator_calculation(pokemon1_type1, pokemon1_type2, pokemon2_type1, pokemon2_type2)
            damage = battle_simulation(attack1, defense2, extra_attack1, modulator)
            hp2 -= damage
        else:
            turn=1
            modulator= moderator_calculation(pokemon2_type1, pokemon2_type2, pokemon1_type1, pokemon1_type2)
            damage = battle_simulation(attack2, defense1, extra_attack2, modulator)
            hp1 -= damage
        if hp1 <= 0:
              return 2
        elif hp2 <= 0:
              return 1
        turn = 1 - turn
    else:
      if(stats1==stats2):
        return 0
      turn = 1 if stats1 > stats2 else 0
      while hp1 > 0 and hp2 > 0:
        if turn != 0:
            modulator= moderator_calculation(pokemon1_type1, pokemon1_type2, pokemon2_type1, pokemon2_type2)
            damage = battle_simulation(attack1, defense2, extra_attack1, modulator)
            hp2 -= damage
        else:
            turn=1
            modulator= moderator_calculation(pokemon2_type1, pokemon2_type2, pokemon1_type1, pokemon1_type2)
            damage = battle_simulation(attack2, defense1, extra_attack2, modulator)
            hp1 -= damage
        if hp1 <= 0:
              return 2
        elif hp2 <= 0:
              return 1
        turn = 1 - turn

    return 0
```

```python
print("Welcome to AI Pokemon Battle Simulation")
switch=int(input("\nPress 1 to start and 0 to Quit"))
while switch == 1:
  pokemon1 = input("\nEnter the first pokemon for battle:").casefold().capitalize()
  pokemon2 = input("\nEnter its Competitor:").casefold().capitalize()
  if pokemon_dataframe[pokemon_dataframe['Name'] == pokemon1].empty or pokemon_dataframe[pokemon_dataframe['Name'] == pokemon2].empty:
    print("One of the pokemon entered are not valid entered are not valid and \nMegas are not valid")
  else:
    winner=battle_stats_outcome_extractor(pokemon1, pokemon2)
    if(winner==1):
      print("\nThe Battle was won by:",pokemon1)
    elif(winner==2):
      print("\nThe Battle was won by:",pokemon2)
    else:
      print("\nThe Victory is hard to determine")
  switch = int(input("\nDo you want to continue \n Enter 1 to continue and 0 to exit  "))
print("\nExiting the simulation.")
```
